# Визуализация

Построим распределение средних значений по каждому из числовых показателей, чтобы сделать вывод об их результатах в целом, в Южной Америке. Для этого построим столбчатую диаграмму.

```python
def average (date):
    return date.mean()
average_values = average(date)
average_values
```

```python
plt.figure(figsize = (10,5))
average_values.plot.bar()
plt.legend(['средние значения показателей'], loc = 'upper left', prop = {'size': 10}) 
plt.title('Столбчатая диаграмма средних значений', color = 'blue', size = 10);
```

>Можно заметить, что самый выской средний показатель у переменных "Schwartz: It is important to this person to be rich" и "Satisfaction with your life". Самые низкие - "Schwartz: It is important to this person to help the people nearby", "How much you trust: Your family". 
>
>Теперь посмотрим на распределения средних для каждой страны, чтобы сравнить эти результаты с общей тенденцией. Сначала сгруппируем данные со средними значениями по каждой стране, после чего построим столбчатую диаграмму.

```python
def country_means(df):
     return df.groupby("Country").mean()
df_country = country_means(df)
del df_country['Year']
df_country
```

```python
plt.figure(figsize = (20,5))
df_country.plot.bar()
plt.title('Столбчатая диаграмма для средних значений каждой страны')
plt.legend(bbox_to_anchor=(1.05, 1), loc = 'upper left')
```

>Итак, можно сделать вывод, что важность семьи, религии и степень доверия людям равны примерно в одинаковой степени в каждой стране. Друзья больше всего ценятся в Перу, больше людей состоит в религиозной организации или считают важным стать богатыми в Бразилии.
